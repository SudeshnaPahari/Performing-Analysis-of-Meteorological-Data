# **performing analysis for testing the (given) Hypothesis**

**Changing the index of Formatted Date into datetime index**

```python
data= pd.read_csv('weatherHistory.csv')
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'], utc=True)
data['Formatted Date']
```

**Making Formatted Date count index**

```python
df=data.set_index('Formatted Date')
df.head()
```

**Resampling data**

```python
df_monthly_mean = df[['Apparent Temperature (C)', 'Humidity']].resample('M').mean()
df_monthly_mean.head()
```

**Plotting the variation in Apparent Temperature and Humidity with time**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,9))
plt.title("Plot of Apparent Temperature and Humidity with time(in year)")
plt.plot(df_monthly_mean['Humidity'],label='Humidity',color='purple', marker='.')
plt.plot(df_monthly_mean['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='red',marker='.')
plt.xlabel("Year",fontsize="16")
plt.legend(loc=2,fontsize=10)
```
